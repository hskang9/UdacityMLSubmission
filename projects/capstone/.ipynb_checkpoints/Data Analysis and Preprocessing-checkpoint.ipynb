# Data Preprocessing

## Replace NaN values with 0

```python
data = data .dropna(axis=0, how="any")
features = data[["NumberOfTime30-59DaysPastDueNotWorse", "NumberOfTimes90DaysLate", "age", "NumberOfTime60-89DaysPastDueNotWorse", "NumberOfDependents"]]
features = features
```

```python
labels = data["SeriousDlqin2yrs"]
```

## Outlier replacement

```python
def percentile_based_outlier(data, threshold=95):
    diff = (100 - threshold) / 2.0
    (minval, maxval) = np.percentile(data, [diff, 100 - diff])
    return ((data < minval) | (data > maxval))


def plotOutlierEachFeature(x):
    fig, axes = plt.subplots(nrows=5)
    nOutliers = []
    for ax, feature in zip(axes, features.keys()):
        tfOutlier = zip(features[feature], percentile_based_outlier(features[feature]))
        nOutliers.append(len([index for (index, bol) in tfOutlier if bol == True]))
        outlierFree = [index for (index, bol) in tfOutlier if bol == True]
        sns.distplot(outlierFree, ax=ax, rug=True, hist=False)
        
    kwargs = dict(y=0.95, x=0.05, ha='left', va='top', size=15)
    axes[0].set_title('NumberOfTime30-59DaysPastDueNotWorse, removed: {r}'.format(r=nOutliers[0]), **kwargs)
    axes[1].set_title('NumberOfTimes90DaysLate, removed: {r}'.format(r=nOutliers[1]), **kwargs)
    axes[2].set_title('age, removed: {r}'.format(r=nOutliers[2]), **kwargs)
    axes[3].set_title('NumberOfTime60-89DaysPastDueNotWorse, removed: {r}'.format(r=nOutliers[3]), **kwargs)
    axes[4].set_title('NumberOfDependents, removed: {r}'.format(r=nOutliers[4]), **kwargs)
    fig.suptitle('Outlier Removed By Percentile-based Method with n={}'.format(len(x)), size=20)
    fig = plt.gcf()
    fig.set_size_inches(15,10)

```
